```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
WC_dict=WC[['Review_Text','Target']]
WC_dict
```

```python
data_set=[]
for i,r in WC_dict.iterrows():
    data_set.append((r['Review_Text'],r['Target']))
```

```python
data_set[:5]
```

```python
print(len(data_set))
```

```python
stemmer=PorterStemmer()
wordnet_lemmatizer=WordNetLemmatizer()
```

```python
def preprocess(document,stem=True):
    document=document.lower()
    words=word_tokenize(document)
    words=[word for word in words if word not in stopwords.words('english')]
    if stem:
        words=[stemmer.stem(word) for word in words]
    else:
        words=[wordnet_lemmatizer.lemmatize(word,pos='v') for word in words]
    document=" ".join(words)
    return document
```

```python
messages_set =[]
for (message, label) in data_set:
    words_filtered=[e.lower() for e in preprocess(message,stem=False).split() if len(e)>=3]
    messages_set.append((words_filtered,label))
```

```python
messages_set[:2]
```

```python
def get_words_in_messages(messages):
    all_words=[]
    for (message,label) in messages:
        all_words.extend(message)
    return all_words
```

# Review Text Preprocessing steps Using Spacy and Gensim 

```python
import re,string,pprint
# spacy for basic preprocessing, optional, can use nltk as well (lemmatisation etc.)
import spacy
```

```python
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
```

```python
import pyLDAvis
import pyLDAvis.gensim
```

```python
df_txt=pd.DataFrame(WC.Review_Text)
```

```python
def sent_to_words(sentences,deacc=True):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence)))
```

```python
data=df_txt.values.tolist()
```

```python
data_words = list(sent_to_words(data))
print(data_words[3])
```

```python
stop_words= stopwords.words('english')+ list(string.punctuation)
```

```python

def remove_stopwords(text):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in text]

data_words_nostops= remove_stopwords(data_words)
```

```python
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
        
    return texts_out
```

```python
nlp=spacy.load('en',disable=['parser','ner'])
data_lemmatized = lemmatization(data_words_nostops, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])
```

# Topic Modelling using LDA

```python
id2word = corpora.Dictionary(data_lemmatized)
```

```python
corpus = [id2word.doc2bow(text) for text in data_lemmatized]
print(corpus[2])
```

```python
# human-readable format of corpus (term-frequency)
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:3]]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=6, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# print the topics
pprint.pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]

doc_lda
```

```python
# coherence score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
